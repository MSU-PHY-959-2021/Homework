# Studying the metalicity of galaxies using CNNs

In this example, we are going to attempt to predict the [metalicity](https://en.wikipedia.org/wiki/Metallicity) of galaxies.  Our goal is to explore convolutional neural networks (CNNs), so we'll use galaxy images from the Sloan Digital Sky Survey.  

```python
# load data sets
train_label_csv = 'metalicity/catalogs/train.csv'
dfTrain = pd.read_csv(train_label_csv, index_col=0)

test_label_csv = 'metalicity/catalogs/test.csv'
dfTest = pd.read_csv(test_label_csv, index_col=0)

# descriptions
print("Description:\n",dfTrain.describe(), "\n")
    
trainLabels = dfTrain.values[:,0].tolist()
testLabels = dfTest.values[:,0].tolist()
```

```python
plt.figure(figsize=(6,3), dpi=150)
plt.hist(dfTrain.oh_p50,bins=25)
plt.xlabel('[O/H] metallicity');
```

```python
list_ds = tf.data.Dataset.list_files('metalicity/train/classOne/*', shuffle=False)
for f in list_ds.take(1):
  print(f.numpy())
```

```python
image_size=(128, 128)
batch_size = 16

train_ds = image_dataset_from_directory(directory='metalicity/train/',
                                        labels=trainLabels,batch_size=batch_size,
                                        image_size=image_size,shuffle=False)

test_ds = image_dataset_from_directory(directory='metalicity/test/',
                                        labels=testLabels,batch_size=batch_size,
                                        image_size=image_size,shuffle=False)
```

```python
plt.figure(figsize=(15, 15))
    
for images, labels in train_ds.take(1):
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(f"Metalicity = {labels[i]:.4f}")
        plt.axis("off")
```

```python
data_augmentation = keras.Sequential(
    [
        layers.experimental.preprocessing.RandomFlip("horizontal"),
        layers.experimental.preprocessing.RandomRotation(0.1),
    ]
)

plt.figure(figsize=(15, 15))
for images, labels in train_ds.take(1):
    for i in range(9):
        augmented_images = data_augmentation(images)
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(augmented_images[2].numpy().astype("uint8"))
        plt.title(f"Metalicity = {labels[2]:.4f}")
        plt.axis("off")
```

```python
#train_ds = train_ds.prefetch(buffer_size=32)
#test_ds = test_ds.prefetch(buffer_size=32)

augmented_train_ds = train_ds.map(
  lambda x, y: (data_augmentation(x, training=True), y))

augmented_test_ds = test_ds.map(
  lambda x, y: (data_augmentation(x, training=False), y))
```

```python
def buildModel(input_shape, final_size, l1_filters, l2_filters, k_size=5, l3_filters=0):
    activate = "elu"

    inputs = keras.Input(shape=input_shape)

    x = layers.experimental.preprocessing.Rescaling(1./255)(inputs)
    print("Input shape: \t",x.shape)
    
    x = layers.experimental.preprocessing.Resizing(final_size,final_size)(x)
    print("Resize shape: \t",x.shape)
    
    x = layers.Conv2D(filters=l1_filters, kernel_size=3)(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation(activate)(x)
    x = layers.MaxPooling2D(pool_size=2)(x)
    print("Post Layer 1 shape: \t",x.shape)
    
    x = layers.Conv2D(filters=l2_filters, kernel_size=k_size)(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation(activate)(x)
    print("Post Layer 2 shape: \t",x.shape)

    if l3_filters > 1:
        x = layers.MaxPooling2D(pool_size=2)(x)
        x = layers.Conv2D(filters=l3_filters, kernel_size=k_size)(x)
        x = layers.BatchNormalization()(x)
        x = layers.Activation(activate)(x)
        print("Post Layer 3 shape: \t",x.shape)


    x = layers.GlobalMaxPooling2D()(x)
    print("Post GlobalMaxPool shape: \t",x.shape)
    print(x.shape)


    outputs = layers.Dense(units=1)(x)

    model = keras.Model(inputs, outputs) 
    print(model.output_shape)
    return model

def trainModel(model, dset, epochs=50, verbose=True, lr=3e-4):
    # We need to compile our model, wherein we define:
    # - The loss function
    # - The SGD optimizer
    # - Any metrics we want the model to track

    optimizer = tf.keras.optimizers.Adam(learning_rate=lr)

    model.compile(loss="MeanSquaredError",optimizer=optimizer,
                  metrics=['MeanAbsolutePercentageError'] )
    
    # Finally, we'll fit our model and then deploy it
    model.fit(dset,
              workers=2,use_multiprocessing=True,
              verbose=verbose,epochs =epochs)
    
    return model
```

```python
aModel = buildModel(input_shape=image_size+(3,), final_size=128, l1_filters=16, l2_filters=32, l3_filters=64)
```

```python
trainModel(aModel, augmented_train_ds, test_ds,epochs = 10, lr=0.01)
trainModel(aModel, augmented_train_ds, test_ds,epochs = 10, lr=0.001)
trainModel(aModel, augmented_train_ds, test_ds,epochs = 10, lr=0.0005)
trainModel(aModel, augmented_train_ds, test_ds,epochs = 10, lr=0.0001)
trainModel(aModel, augmented_train_ds, test_ds,epochs = 10, lr=0.00005)
```

```python
predictionsTrain = aModel.predict(train_ds)

predictionsTest = aModel.predict(test_ds)
print("One prediction and its true value: ",predictionsTrain[0], trainLabels[0])
```

```python
fig = plt.figure(figsize=(10,8))
plt.scatter(trainLabels,predictionsTrain)
plt.xlim(8, 10)
plt.ylim(8, 10)
plt.ylabel("Z Predicted")
plt.xlabel("Z True")
plt.title("Training Data")
plt.plot(plt.xlim(), plt.ylim(), ls='--', c='k');
plt.show()

fig = plt.figure(figsize=(10,8))
plt.scatter(testLabels,predictionsTest)
plt.xlim(8, 10)
plt.ylim(8, 10)
plt.ylabel("Z Predicted")
plt.xlabel("Z True")
plt.title("Testing Data")
plt.plot(plt.xlim(), plt.ylim(), ls='--', c='k');
plt.show()
```

```python
ratioTrain = np.copy(predictionsTrain)
for idx, label in  enumerate(trainLabels):
    ratioTrain[idx] /= label
print("Training mean/rms: ",ratioTrain.mean(), ratioTrain.std())
plt.hist(ratioTrain,bins=100)
plt.show()

ratioTest = np.copy(predictionsTest)
for idx, label in  enumerate(testLabels):
    ratioTest[idx] /= label
print("Testing mean/rms: ",ratioTest.mean(), ratioTest.std())
plt.hist(ratioTest,bins=100)
plt.show()
```
